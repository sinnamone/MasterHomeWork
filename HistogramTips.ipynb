```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
```

```python
df = pd.read_csv("/Users/simone/Documents/GitHub/MasterHomeWork/hq_risto_info.csv",sep="\t",header=0)
```

```python
def histo_plot(df,ristorante):
    plt.rcParams['figure.figsize'] = [10, 5]
    colors = ['green', 'orange', 'blue','red']
    label=['']
    fig, ax = plt.subplots()
    df['new_title']
    counts, bins, patches = ax.hist(df['distanza_text'], facecolor='yellow', edgecolor='gray',bins=40)
    # Change the colors of bars at the edges...
    twentyfifth,fifth, seventyfifth,one = np.percentile(df['distanza_text'], [25,50, 75,98])
    for patch, rightside, leftside in zip(patches, bins[1:], bins[:-1]):
        if rightside < twentyfifth:
            patch.set_facecolor('green')
        if (rightside > twentyfifth) & (rightside  < fifth):
            patch.set_facecolor('orange')
        if (rightside > fifth) & (rightside  < seventyfifth):
            patch.set_facecolor('blue')
        elif (rightside > seventyfifth):
            patch.set_facecolor('red')
    plt.axvline(int(df.loc[df['new_title'].str.contains(ristorante)==True]['distanza_text']), color='k', linestyle='dashed', linewidth=1)
    #plt.text(df['distanza_text'].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x.mean()))
    plt.xlabel(' Distanza (Metri)', fontsize=18)
    plt.ylabel(' Frequenza ', fontsize=18)
    
    green_patch = mpatches.Patch(color='green', label='Tips= 5%')    
    orange_patch = mpatches.Patch(color='orange', label='Tips= 10%')    
    blue_patch = mpatches.Patch(color='blue', label='Tips= 15%')
    red_patch = mpatches.Patch(color='red', label='Tips= 20%')

    ax.legend(handles=[green_patch,orange_patch,blue_patch,red_patch])
    plt.subplots_adjust(bottom=0.15)
    plt.show()
```

```python
histo_plot(df,"poke house ticinese")
```
